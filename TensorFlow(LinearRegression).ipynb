```
import pandas as pd
import numpy as np
import tensorflow as tf
from matplotlib import  pyplot as plt
from google.colab import files
from google.colab import drive
import io
```

```
#read file from local 
df = files.upload()
#upload files from local -> Dataframe
data = pd.read_csv(io.BytesIO(df["Income1.csv"]))
```

```
print(data)
```

```
feature = data.Education
lable = data.Income
plt.scatter(feature,lable)
```

```
#Building a basic model
model = tf.keras.Sequential() # Empty model
```

```
#Add layer to model
model.add(tf.keras.layers.Dense(units=1,input_shape = (1,)))
#units = 1 => output only 1 cell(income)
#input_sharpe = ?
#input_sharpe(1,) => input data = 1 dim(Eductio) 
```

```
model.summary()
```

(None , 1) None mean how much input simple will we input(ex:100 200 etc) no need to consider

And 1 mean output 1 dim

param = how many variable
parm 2 = Ax + B (A(weight) and B(Bais) variale)

```
model.compile(optimizer="adam",loss = "mean_squared_error")
#adam = find the minimum value in the curse
#loss= a calculate loss method = (y-fx)^2 *1/
```

```
#Trainning Model:
history = model.fit(feature,lable,batch_size=30,epochs=5000)
#epochs  = trainning time
#batch_size = feature data size/number of your input data
```

```
#Using Trainning model to predicat our input
#Is that predict as we expected?
#26(label) and 32(perdicated)
model.predict(feature)
```

```
model.predict(pd.Series([50]))
#predict the income of 50years later
```
